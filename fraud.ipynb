```python
def predict_fraudtype(amount=1262.77,card_type="Rupay",location="Bangalore",purchase_category="Digital",customer_age=28,hour=22.0,day=24.0,weekday=4.0,month=11.0):
    lst=[]
    lst=lst+[amount]
    print(card_type)   
    if card_type=="MasterCard":
        lst=lst+[1]
    elif card_type=="Rupay":
        lst=lst+[2]
    elif card_type=="Visa":
        lst=lst+[3]
    location=list(lb_LN.transform([location]))
    lst=lst+location
    print(lst)

    if purchase_category=="Digital":
        lst=lst+[1]
    elif purchase_category=="POS":
        lst=lst+[2]
    lst=lst+[customer_age]
    lst=lst+[hour]
    lst=lst+[day]
    lst=lst+[weekday]
    lst=lst+[month]
    print(lst)
    result=model.predict([lst])   
    if result==[1]:
        return "Identity theft"
    elif result==[2]:
        return "Malware"
    elif result==[3]:
        return "Payment card fraud"
    elif result==[4]:
        return "phishing"
    elif result==[5]:
        return "scam"
    else:
        return"no fraud"
print(predict_fraudtype())
    
            
```
